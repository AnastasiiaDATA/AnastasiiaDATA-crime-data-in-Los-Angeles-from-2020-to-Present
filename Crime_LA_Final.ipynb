```python
## Import Required Libraries
We import the necessary libraries: `pandas` for data manipulation and `matplotlib.pyplot` for visualization.

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

## Understand and Clean Data

```python
# Load the Dataset
We read the crime dataset from a CSV file and display the first few rows to understand its structure.

# Improved: Read the data
crime = pd.read_csv('Crime_Data_from_2020_to_Present.csv')
crime.head()
```

```python
crime.columns
```

```python
crime.shape
```

```python
crime.dtypes
```

```python
# Check Nulls
crime.isna().sum
```

```python
df = crime.drop(labels =['modus_operandi', 'weapon_code', 'weapon_description',
'crime_code_2', 'crime_code_3', 'crime_code_4', 'cross_street'], axis = 1)
df
```

```python
df.sort_values(by = 'date_reported')
```

```python
df.dtypes
```

```python
# Update datatypes
df['date_reported'] = pd.to_datetime(df['date_reported'], format = '%Y-%m-%d')
df['date_occurred'] = pd.to_datetime(df['date_occurred'], format='%Y-%m-%d %H:%M:%S')
df
```

```python
# Check Dataset Information
This provides an overview of the dataset, including column names, data types, and missing values.

df.info()
```

```python
df.describe(include =['float64', 'int64'])
```

```python
df['victim_age'].value_counts()
```

```python
df_age = df[df['victim_age'] > 0]
df_age
```

```python
df_age['victim_age'].value_counts()
```

```python
df['year'] = df['date_occurred'].dt.year
df['month'] = df['date_occurred'].dt.month
df.sort_values(by = 'date_reported')
```

## Feature Understanfing

```python
df['month'].value_counts().sort_index().plot()
```

```python
df['year'].value_counts().sort_index().plot(kind='bar', edgecolor = 'black',  color = 'skyblue')
```

```python
df = df[~((df['month'] == 12) & (df['year'] == 2023))]
```

```python
df.sort_values(by = 'date_occurred')
```

```python
# Crime Count by Category
We group data by crime type and count occurrences to see the most common crimes.

crime_over_time = df.groupby(['year', 'month']).size().plot()
crime_over_time
```

```python
# Visualizing Crime Data
We create visualizations to better understand crime trends.

plt.hist(df_age['victim_age'],edgecolor = 'black',  color = 'skyblue', bins = 15)
```

```python
df['date_occurred_no_time'] = df['date_occurred'].dt.date
```

```python
# Crime Count by Category
We group data by crime type and count occurrences to see the most common crimes.

crime_count_per_day = df.groupby(['date_occurred_no_time', 'crime_description']).size()
```

```python
top_20_crimes = crime_count_per_day.groupby('crime_description').mean().reset_index(name = 'average_daily_count').sort_values(by =  'average_daily_count', ascending = False).head(20)
```

```python
top_20_crimes_sorted = top_20_crimes.sort_values(by = 'average_daily_count', ascending = True)
```

```python
# Visualizing Crime Data
We create visualizations to better understand crime trends.

plt.barh(top_20_crimes['crime_description'], top_20_crimes_sorted['average_daily_count'])
```

```python
# Crime Count by Category
# We group data by crime type and count occurrences to see the most common crimes.

plt.figure(figsize = (12,8))

#  Plot 1 - Line Chart
plt.subplot(2,2,1)
crime_plot = df.groupby(['year', 'month']).size().plot(linewidth = 2)
crime_plot.set_xlabel('Year - Month', fontsize = 11)
crime_plot.set_title('Crime over Time')

# Plot 2 - Histogram
plt.subplot(2,2,2)
plt.hist(df_age['victim_age'],edgecolor = 'black',  color = 'orange', bins = 15)
plt.tight_layout()
plt.xlabel('Age', fontsize = 11)
plt.title('Distribution of Victime Age')

# Plot 3 - Horizontal Bar
plt.figure()
plt.barh(top_20_crimes['crime_description'], top_20_crimes_sorted['average_daily_count'], color = 'green')
plt.title('Top 20 Crimes')
```
